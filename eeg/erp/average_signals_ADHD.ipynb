```python
import numpy as np
import scipy.io as sio
import h5py

# File paths
d5_file_path = '/content/drive/MyDrive/EEG_dataset/dataset/d5.mat'
output_file_path = '/content/drive/MyDrive/EEG_dataset/dataset/last_1840_of_d5.mat'

def load_mat_file(file_path, key=None):
    """
    Load a .mat file and extract the data using the specified key.
    Automatically infer the key if not provided.
    """
    try:
        with h5py.File(file_path, 'r') as f:  # For v7.3 .mat files
            if key is None:
                key = list(f.keys())[0]  # Automatically infer key
            data = np.array(f[key])
    except OSError:  # Handle older .mat files
        mat_data = sio.loadmat(file_path)
        if key is None:
            key = [k for k in mat_data.keys() if not k.startswith("__")][0]
        data = mat_data[key]
    return data

# Load the d5 file
d5_data = load_mat_file(d5_file_path, key='d5')  # Replace 'd5' with the correct key if necessary

# Extract the last 1840 columns
last_1840_columns = d5_data[:, :, -1840:]

# Save the extracted data to a new .mat file
sio.savemat(output_file_path, {'last_1840_columns': last_1840_columns})

print(f"Last 1840 columns of d5 saved successfully to {output_file_path}")
```

```python
import numpy as np
import scipy.io as sio

# File paths
ystim_file_path = '/content/drive/MyDrive/EEG_dataset/y_stim_subset_23161_to_33902.mat'
combined_eeg_file_path = '/content/drive/MyDrive/EEG_dataset/ADHD_Files_combined.mat'

# Load y_stim subset
ystim_data = sio.loadmat(ystim_file_path)['y_stim_subset_23161_to_33902']  # Replace with actual key if different
participants = ystim_data[0, :]  # Flatten to get participant IDs per trial

# Load combined EEG data
combined_data = sio.loadmat(combined_eeg_file_path)['combined_data']  # Replace with actual key if different

# Step 1: Calculate the number of trials per participant
unique_participants, trial_counts = np.unique(participants, return_counts=True)
participant_trial_mapping = dict(zip(unique_participants, trial_counts))

# Print the mapping
print("Participant ID → Number of Trials")
for participant_id, count in participant_trial_mapping.items():
    print(f"Participant {participant_id}: {count} trials")

# Step 2: Map trials and compute average EEG signal for each participant
start_index = 0  # Initialize starting index for trials
for participant_id in unique_participants:
    # Number of trials for the participant
    num_trials = participant_trial_mapping[participant_id]

    # Extract trials for this participant from the combined EEG data
    participant_trials = combined_data[:, :, start_index:start_index + num_trials]

    # Compute the average EEG signal for this participant (mean across trials)
    average_signal = np.mean(participant_trials, axis=2)  # Shape: [timepoints, channels]

    # Transpose the signal to have shape [channels, timepoints]
    average_signal = average_signal.T  # Shape: [channels, timepoints]

    # Save the average signal for this participant to a .mat file
    output_path = f'/content/drive/MyDrive/EEG_dataset/ADHD_ERPs/average_signal_participant_{participant_id}.mat'
    sio.savemat(output_path, {'average_signal': average_signal})

    # Print progress
    print(f"Average signal for Participant {participant_id} saved to {output_path}")

    # Update start index for the next participant
    start_index += num_trials
```

```python
import numpy as np
import scipy.io as sio
import h5py

# File paths
d1_file_path = '/content/drive/MyDrive/EEG_dataset/dataset/d1.mat'
d2_file_path = '/content/drive/MyDrive/EEG_dataset/dataset/d2.mat'
first_129_of_d3_file_path = '/content/drive/MyDrive/EEG_dataset/dataset/first_129_of_d3.mat'
output_file_path = '/content/drive/MyDrive/EEG_dataset/control_Files_combined.mat'

def load_mat_file(file_path, key=None):
    """
    Load a .mat file and extract the data using the specified key.
    Automatically infer the key if not provided.
    """
    try:
        with h5py.File(file_path, 'r') as f:  # For v7.3 .mat files
            if key is None:
                key = list(f.keys())[0]  # Automatically infer key
            data = np.array(f[key])
    except OSError:  # Handle older .mat files
        mat_data = sio.loadmat(file_path)
        if key is None:
            key = [k for k in mat_data.keys() if not k.startswith("__")][0]
        data = mat_data[key]
    return data

# Load the three files
d1_data = load_mat_file(d1_file_path, key='d1')  # Replace 'd1' with the correct key if necessary
d2_data = load_mat_file(d2_file_path, key='d2')  # Replace 'd2' with the correct key if necessary
first_129_of_d3_data = load_mat_file(first_129_of_d3_file_path, key='first_129_columns')  # Adjust key as needed

# Ensure dimensions align for concatenation
assert d1_data.shape[:2] == d2_data.shape[:2] == first_129_of_d3_data.shape[:2], "Mismatch in row or channel dimensions"

# Concatenate along the third axis (assuming trials are along the third axis)
combined_data = np.concatenate((d1_data, d2_data, first_129_of_d3_data), axis=2)

# Save the combined data into a new .mat file
sio.savemat(output_file_path, {'combined_data': combined_data})

print(f"Combined data saved successfully to {output_file_path}")
```

```python
import os
import numpy as np
import scipy.io as sio

# File paths
ystim_file_path = '/content/drive/MyDrive/EEG_dataset/y_stim_subset_1_to_10129.mat'
combined_eeg_file_path = '/content/drive/MyDrive/EEG_dataset/control_Files_combined.mat'
output_dir = '/content/drive/MyDrive/EEG_dataset/CONTROL_ERPs/'

# Ensure output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Load y_stim subset
ystim_data = sio.loadmat(ystim_file_path)['y_stim_subset']
participants = ystim_data[0, :]  # Flatten to get participant IDs per trial

# Load combined EEG data
combined_data = sio.loadmat(combined_eeg_file_path)['combined_data']

# Step 1: Calculate the number of trials per participant
unique_participants, trial_counts = np.unique(participants, return_counts=True)
participant_trial_mapping = dict(zip(unique_participants, trial_counts))

# Print the mapping
print("Participant ID → Number of Trials")
for participant_id, count in participant_trial_mapping.items():
    print(f"Participant {int(participant_id)}: {count} trials")

# Step 2: Map trials and compute average EEG signal for each participant
start_index = 0  # Initialize starting index for trials
for participant_id in unique_participants:
    num_trials = participant_trial_mapping[participant_id]

    # Extract trials for this participant from the combined EEG data
    participant_trials = combined_data[:, :, start_index:start_index + num_trials]

    # Compute the average EEG signal for this participant
    average_signal = np.mean(participant_trials, axis=2)

    # Transpose the signal
    average_signal = average_signal.T

    # Save the average signal
    output_path = os.path.join(output_dir, f'average_signal_participant_control_{int(participant_id)}.mat')
    sio.savemat(output_path, {'average_signal': average_signal})

    print(f"Average signal for Participant {int(participant_id)} saved to {output_path}")

    start_index += num_trials
```
