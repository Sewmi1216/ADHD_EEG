```python
import os
import mne
import matplotlib.pyplot as plt

# Set the directory path to your folder containing the ERP files
folder_path = '/content/drive/MyDrive/dataset/adhderps'
output_folder = '/content/drive/MyDrive/dataset/plots/erp_plots'  # Folder where you want to save the plots

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# List all ERP files in the directory (ending with '_erp.fif')
files = [f for f in os.listdir(folder_path) if f.endswith('_erp.fif')]

# Iterate through each file
for file in files:
    # Load the Evoked data from the file (condition=0 is specified)
    file_path = os.path.join(folder_path, file)
    evoked_data = mne.read_evokeds(file_path, condition=0)  # Single Evoked object for each file

    print(f"Plotting ERP for file: {file}")

    # Plot ERP for each channel
    for channel_name in evoked_data.info['ch_names']:
        # Select the channel using the 'picks' argument
        fig = evoked_data.plot(picks=[channel_name], show=False)

        # Set the title for the plot using matplotlib
        fig.suptitle(f"ERP for {file}, Channel {channel_name}", fontsize=16)

        # Define the path to save the plot
        save_path = os.path.join(output_folder, f"{file}_{channel_name}_erp.png")

        # Save the plot
        plt.savefig(save_path)

        # Close the plot to avoid memory issues
        plt.close()
```
